```python
!pip install matplotlib

import matplotlib.pyplot as plt

# Define the graphing function
def make_graph(data, column='Close', title='Stock Price Over Time'):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data[column], label=column)
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Download historical data for Tesla
tsla = yf.Ticker("TSLA")
tsla_data = tsla.history(period="max")

# Display the first few rows
print(tsla_data.head())

# Call make_graph() to visualize
make_graph(tsla_data, column='Close', title='Tesla Closing Price Over Time')

```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for GME
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Display the first few rows of data
print(gme_data.head())

# Plot the closing price over time
def make_graph():
    plt.figure(figsize=(12, 6))
    plt.plot(gme_data.index, gme_data['Close'], label='GME Closing Price')
    plt.title("GME Stock Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

make_graph()
```
