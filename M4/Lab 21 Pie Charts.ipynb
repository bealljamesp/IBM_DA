### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# Step 1: Split and count each database
all_databases = df['DatabaseWantToWorkWith'].dropna().str.split(';')
flat_list = [db.strip() for sublist in all_databases for db in sublist]
database_counts = pd.Series(flat_list).value_counts()

# Step 2: Get the top 5 most-desired databases
top5 = database_counts.head(5)

# Step 3: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Most-Desired Databases')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.tight_layout()
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.

# Step 1: Split and count each developer type
# Assume 'DevType' column contains entries like: "Data scientist;Data analyst;Back-end developer"
all_roles = df['DevType'].dropna().str.split(';')
flat_roles = [role.strip() for roles in all_roles for role in roles]
role_counts = pd.Series(flat_roles).value_counts()

# Step 2: Select the top 5 roles
top5_roles = role_counts.head(5)

# Step 3: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(top5_roles, labels=top5_roles.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Developer Roles')
plt.axis('equal')  # Ensures pie is a circle
plt.tight_layout()
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
# The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.

# Step 1: Count the occurrences of each operating system
os_counts = df['OpSysProfessional use'].dropna().value_counts()

# Step 2: Get the top 5 operating systems
top5_os = os_counts.head(5)

# Step 3: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(top5_os, labels=top5_os.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Operating Systems Used Professionally')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.tight_layout()
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
# pie chart with LanguageHaveWorkedWith column

# Step 1: Split and count each programming language
all_languages = df['LanguageHaveWorkedWith'].dropna().str.split(';')
flat_languages = [lang.strip() for sublist in all_languages for lang in sublist]
language_counts = pd.Series(flat_languages).value_counts()

# Step 2: Get the top 5 most-used programming languages
top5_languages = language_counts.head(5)

# Step 3: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(top5_languages, labels=top5_languages.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Most-Used Programming Languages')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.tight_layout()
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
# pie chart for NEWCollabToolsHaveWorkedWith column

# Step 1: Split and count each collaboration tool
all_tools = df['NEWCollabToolsHaveWorkedWith'].dropna().str.split(';')
flat_tools = [tool.strip() for sublist in all_tools for tool in sublist] # remove leading/trailing spaces
tool_counts = pd.Series(flat_tools).value_counts()

# Step 2: Get the top 5 most-used collaboration tools
top5_tools = tool_counts.head(5)

# Step 3: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(top5_tools, labels=top5_tools.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Most-Used Collaboration Tools')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.tight_layout()
plt.show()
```

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages
