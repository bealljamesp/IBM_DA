### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
# Duplicates based on MainBranch only
duplicate_MainBranch = df[df.duplicated(subset=['MainBranch'])]
print("Duplicates based on MainBranch:")
print(duplicate_MainBranch)

# Duplicates based on Employment only
duplicate_Employment = df[df.duplicated(subset=['Employment'])]
print("\n Duplicates based on Employment:")
print(duplicate_Employment)

# Duplicates based on RemoteWork only
duplicate_RemoteWork = df[df.duplicated(subset=['RemoteWork'])]
print("\n Duplicates based on RemoteWork:")
print(duplicate_RemoteWork)
```

```python
print("MainBranch duplicates:", df.duplicated(subset=['MainBranch']).sum())
print("Employment duplicates:", df.duplicated(subset=['Employment']).sum())
print("RemoteWork duplicates:", df.duplicated(subset=['RemoteWork']).sum())
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
import matplotlib.pyplot as plt

# Function to plot duplicate distributions
def plot_duplicates(df, column_name):
    counts = df[column_name].value_counts()
    duplicates = counts[counts > 1].head(5)  # Keep only duplicated values

    if duplicates.empty:
        print(f"No duplicates found in column: {column_name}")
        return

    plt.figure(figsize=(8, 4))
    duplicates.plot(kind='bar')
    plt.title(f"Distribution of Duplicates in '{column_name}'")
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Plot duplicates for each column
plot_duplicates(df, 'MainBranch')
plot_duplicates(df, 'Employment')
plot_duplicates(df, 'RemoteWork')
```
