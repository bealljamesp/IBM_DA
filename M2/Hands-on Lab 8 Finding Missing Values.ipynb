### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
missing_values = df.isna().sum()
print(missing_values)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
# Set figure size
plt.figure(figsize=(12, 6))

# Plot heatmap where True = missing (NaN), False = present
sns.heatmap(df.isna(), 
            cmap='viridis', 
            cbar=False, 
            yticklabels=False)

plt.title("Heatmap of Missing Values by Column")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 6))

sns.heatmap(df.isna(), 
            cmap='viridis', 
            cbar=False, 
            yticklabels=False)

plt.title("Heatmap of Missing Values by Column", fontsize=10)  # smaller title
plt.xlabel("Columns", fontsize=8)
plt.ylabel("Rows", fontsize=8)

# Make tick labels smaller
plt.xticks(fontsize=6, rotation=45, ha='right')
plt.yticks(fontsize=6)

plt.tight_layout()
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
missing_employment = df['Employment'].isna().sum()
print(f"Number of missing values in 'Employment': {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
top_employment = df['Employment'].value_counts().head(5)
print("Top 5 Employment categories:")
print(top_employment)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
employment_counts = df['Employment'].value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(employment_counts, 
        labels=employment_counts.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        shadow=True)

plt.title("Employment Distribution")
plt.axis('equal')  # Ensures the pie is a circle
plt.tight_layout()
plt.show()
```
